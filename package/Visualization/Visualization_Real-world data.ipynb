```python
fig, ax = plt.subplots(figsize=(10, 5))

fs = 20
width = 1/6

xticks = [i + width * 2 for i in range(1, 4)]
yticks = [0, 0.2, 0.4, 0.6]

b1 = ax.bar([i + 0 * width for i in range(1, 4)], [0.62, 0.6, 0.64], width, hatch='//', fc='none', ec=['black'] * 3)
b2 = ax.bar([i + 1 * width for i in range(1, 4)], [0.3, 0.29, 0.3], width, hatch='++', fc='none', ec=['black'] * 3)
b3 = ax.bar([i + 2 * width for i in range(1, 4)], [0.21, 0.27, 0.23], width, hatch='--', fc='none', ec=['black'] * 3)
b4 = ax.bar([i + 3 * width for i in range(1, 4)], [0.21, 0.26, 0.23], width, hatch='**', fc='none', ec=['black'] * 3)
b5 = ax.bar([i + 4 * width for i in range(1, 4)], [0.12, 0.13, 0.11], width, hatch='..', fc='none', ec=['black'] * 3)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels(('precision', 'recall', 'nDCG'))
ax.legend((b1[0], b2[0], b3[0], b4[0], b5[0]), ('RCAE2E', 'RCA', 'gRank', 'mRank', 'LBP'), \
          prop={'size': fs}, edgecolor='black', bbox_to_anchor=(1.3, 1.035))
ax.tick_params(labelsize = fs)
```
